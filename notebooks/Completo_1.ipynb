```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score
import matplotlib.pyplot as plt
from tqdm import tqdm
from collections import defaultdict
import joblib
import warnings

# Sopprimi avvisi non critici
warnings.filterwarnings("ignore", category=UserWarning)

# 1) Load & filter data
print("Caricamento dati...")
df = pd.read_csv('/home/jupyter-leosky01/ATP_pred/data/atp_tennis_cleaned.csv', parse_dates=['Date'])
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
start, end = pd.to_datetime('2020-01-01'), pd.to_datetime('2026-02-08')
df = df[(df['Date'] >= start) & (df['Date'] <= end)].copy().sort_values('Date')

# 2) Feature Engineering Avanzato
print("Creazione feature...")
# Calcola le serie di vittorie
df['p1_win'] = (df['Winner'] == df['Player_1']).astype(int)
df['p2_win'] = (df['Winner'] == df['Player_2']).astype(int)

# Calcola le statistiche mobili per ogni giocatore
print("Calcolo statistiche mobili...")
player_stats = []
for player in tqdm(set(df['Player_1']).union(set(df['Player_2'])), desc="Elaborazione giocatori"):
    player_matches = pd.concat([
        df[df['Player_1'] == player][['Date', 'p1_win']].rename(columns={'p1_win': 'win'}),
        df[df['Player_2'] == player][['Date', 'p2_win']].rename(columns={'p2_win': 'win'})
    ]).sort_values('Date')
    
    player_matches['win_rate'] = player_matches['win'].transform(
        lambda x: x.rolling(window=10, min_periods=1).mean().shift(1)
    )
    player_matches['streak'] = player_matches['win'].groupby(
        (player_matches['win'] != player_matches['win'].shift()).cumsum()
    ).cumcount() + 1
    player_matches['streak'] = player_matches['streak'] * player_matches['win']
    
    player_stats.append(player_matches[['Date', 'win_rate', 'streak']].assign(player=player))

player_stats = pd.concat(player_stats)

# 3) Merge delle statistiche avanzate
print("Merge statistiche...")
def merge_player_stats(df_row, stats_df, player_col, date_col):
    stats = stats_df[
        (stats_df['player'] == df_row[player_col]) & 
        (stats_df['Date'] < df_row['Date'])
    ].sort_values('Date', ascending=False)
    
    if stats.empty:
        return 0.5, 0  # Default se non ci sono partite precedenti
    else:
        return stats.iloc[0]['win_rate'], stats.iloc[0]['streak']

# Applica la funzione di merge
df[['p1_win_rate', 'p1_streak']] = df.apply(
    lambda x: merge_player_stats(x, player_stats, 'Player_1', 'Date'), 
    axis=1, result_type='expand'
)
df[['p2_win_rate', 'p2_streak']] = df.apply(
    lambda x: merge_player_stats(x, player_stats, 'Player_2', 'Date'), 
    axis=1, result_type='expand'
)

# 4) Head-to-Head
print("Calcolo testa a testa...")
h2h = defaultdict(lambda: {'p1_wins': 0, 'matches': 0})
h2h_features = []

for _, row in tqdm(df.iterrows(), total=len(df), desc="Calcolo H2H"):
    p1, p2 = row['Player_1'], row['Player_2']
    # Crea una chiave che mantiene l'ordine originale (non ordinato alfabeticamente)
    key = (p1, p2)
    
    # Aggiorna le statistiche H2H
    if key not in h2h:
        # Se non esiste ancora questa combinazione, inizializza
        h2h[key] = {'p1_wins': 0, 'matches': 0}
    
    if h2h[key]['matches'] > 0:
        # Calcola il win rate per P1 rispetto a P2 (mantenendo l'ordine originale)
        h2h_win_rate = h2h[key]['p1_wins'] / h2h[key]['matches']
    else:
        h2h_win_rate = 0.5  # default se non ci sono match precedenti
    
    # Aggiungi le feature H2H
    h2h_features.append({
        'h2h_win_rate': h2h_win_rate,  # Non invertire in base all'ordine alfabetico
        'h2h_matches': h2h[key]['matches']
    })
    
    # Aggiorna h2h per il prossimo match
    if row['Winner'] == p1:
        h2h[key]['p1_wins'] += 1
    h2h[key]['matches'] += 1

df = pd.concat([df, pd.DataFrame(h2h_features)], axis=1)

# 5) Trasformazione in feature differenziali e assolute (invarianti rispetto all'ordine)
print("Creazione feature invarianti...")
# Feature differenziali
df['rank_diff'] = df['Rank_1'] - df['Rank_2']  # Differenza nei ranking
df['odds_ratio'] = df['Odd_1'] / df['Odd_2']   # Rapporto tra quote
df['win_rate_diff'] = df['p1_win_rate'] - df['p2_win_rate']  # Differenza nei tassi di vittoria
df['streak_diff'] = df['p1_streak'] - df['p2_streak']  # Differenza negli streak

# Feature assolute
df['min_rank'] = np.minimum(df['Rank_1'], df['Rank_2'])  # Ranking del giocatore più forte
df['max_rank'] = np.maximum(df['Rank_1'], df['Rank_2'])  # Ranking del giocatore più debole
df['min_odds'] = np.minimum(df['Odd_1'], df['Odd_2'])    # Quota più bassa
df['max_odds'] = np.maximum(df['Odd_1'], df['Odd_2'])    # Quota più alta
df['max_win_rate'] = np.maximum(df['p1_win_rate'], df['p2_win_rate'])  # Win rate più alto
df['min_win_rate'] = np.minimum(df['p1_win_rate'], df['p2_win_rate'])  # Win rate più basso
df['max_streak'] = np.maximum(df['p1_streak'], df['p2_streak'])  # Streak più lungo
df['min_streak'] = np.minimum(df['p1_streak'], df['p2_streak'])  # Streak più corto

# 6) Split dei dati PRIMA della normalizzazione
print("Split dei dati...")
n = len(df)
idx = np.arange(n)
np.random.seed(42)
np.random.shuffle(idx)
t0, t1 = int(0.7*n), int(0.85*n)
train_i, val_i, test_i = idx[:t0], idx[t0:t1], idx[t1:]

# Definizione delle nuove colonne
num_cols = ['rank_diff', 'odds_ratio', 'win_rate_diff', 'streak_diff',
            'min_rank', 'max_rank', 'min_odds', 'max_odds',
            'max_win_rate', 'min_win_rate', 'max_streak', 'min_streak',
            'h2h_win_rate', 'h2h_matches']
cat_cols = ['Surface', 'Tournament']

# 7) Preparazione dei dati
print("Preparazione dati per il modello...")
X_num = df[num_cols].fillna(0).values
X_cat = OneHotEncoder(handle_unknown='ignore', sparse_output=False)\
        .fit_transform(df[['Surface','Tournament']].fillna('NA'))

# Normalizza SOLO sui dati di training
scaler = StandardScaler()
X_num_train = scaler.fit_transform(X_num[train_i])
X_num = scaler.transform(X_num)  # Applica la stessa trasformazione a tutti i dati

X = np.hstack([X_num, X_cat])
y = df['p1_win'].fillna(0).astype(int).values  # Target: player1 vince? (con pulizia dei dati)

# 8) Dataset e DataLoader
class TDataset(Dataset):
    def __init__(self, X, y, i):
        self.X = torch.tensor(X[i], dtype=torch.float32)
        self.y = torch.tensor(y[i], dtype=torch.float32).unsqueeze(1)
        
    def __len__(self): 
        return len(self.y)
    
    def __getitem__(self, idx): 
        return self.X[idx], self.y[idx]

train_ds = TDataset(X, y, train_i)
val_ds = TDataset(X, y, val_i)
test_ds = TDataset(X, y, test_i)

train_loader = DataLoader(train_ds, batch_size=128, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=128)
test_loader = DataLoader(test_ds, batch_size=128)

# 9) Modello migliorato
class EnhancedMLP(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.bn1 = nn.BatchNorm1d(in_dim)
        self.net = nn.Sequential(
            nn.Linear(in_dim, 512), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4929775),
            nn.BatchNorm1d(512),
            
            nn.Linear(512, 128), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4220651),
            nn.BatchNorm1d(128),
            
            nn.Linear(128, 64),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.1701431),
            
            nn.Linear(64, 1)
        )
        
    def forward(self, x):
        return self.net(self.bn1(x))

# 10) Inizializzazione e training
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = EnhancedMLP(X.shape[1]).to(device)

# Calcola i pesi delle classi direttamente (più sicuro)
pos_samples = np.sum(y[train_i])
neg_samples = len(y[train_i]) - pos_samples
pos_weight = ((neg_samples / pos_samples) + 9.05) if pos_samples > 0 else 1.0

print(f"Campioni positivi: {pos_samples}, Campioni negativi: {neg_samples}")
print(f"Peso positivo: {pos_weight}")

# Usa i pesi nella loss function
criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor(pos_weight, device=device))
optimizer = optim.Adam(model.parameters(), lr=9.00064e-05, weight_decay=8.196026e-05)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'max', patience=5, factor=0.28)

# Funzioni di training e validazione
def train_epoch(model, loader, criterion, optimizer, device):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []
    
    for inputs, labels in tqdm(loader, desc="Training"):
        inputs, labels = inputs.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
        all_preds.extend(torch.sigmoid(outputs).detach().cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(loader)
    auc = roc_auc_score(all_labels, all_preds)
    return avg_loss, auc

def evaluate(model, loader, criterion, device):
    model.eval()
    total_loss = 0
    all_preds, all_labels = [], []
    
    with torch.no_grad():
        for inputs, labels in tqdm(loader, desc="Evaluating"):
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            
            total_loss += loss.item()
            all_preds.extend(torch.sigmoid(outputs).cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    
    avg_loss = total_loss / len(loader)
    auc = roc_auc_score(all_labels, all_preds)
    accuracy = accuracy_score(all_labels, np.array(all_preds) > 0.5)
    return avg_loss, auc, accuracy, all_preds, all_labels

# 11) Addestramento con early stopping
print("Inizio addestramento...")
n_epochs = 100
best_val_auc = 0
patience = 30
patience_counter = 0
best_model = None

train_losses, val_losses = [], []
train_aucs, val_aucs = [], []

for epoch in range(n_epochs):
    print(f"Epoch {epoch+1}/{n_epochs}")
    
    # Training
    train_loss, train_auc = train_epoch(model, train_loader, criterion, optimizer, device)
    train_losses.append(train_loss)
    train_aucs.append(train_auc)
    
    # Validation
    val_loss, val_auc, val_acc, _, _ = evaluate(model, val_loader, criterion, device)
    val_losses.append(val_loss)
    val_aucs.append(val_auc)
    
    print(f"Train Loss: {train_loss:.4f}, Train AUC: {train_auc:.4f}")
    print(f"Val Loss: {val_loss:.4f}, Val AUC: {val_auc:.4f}, Val Acc: {val_acc:.4f}")
    
    # Scheduler step
    scheduler.step(val_auc)
    
    # Early stopping
    if val_auc > best_val_auc:
        best_val_auc = val_auc
        patience_counter = 0
        best_model = model.state_dict().copy()
        torch.save(best_model, '/home/jupyter-leosky01/ATP_pred/best_model_invariant.pt')
        print(f"Nuovo miglior modello salvato! AUC: {val_auc:.4f}")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break

# 12) Carica il miglior modello e test
print("Valutazione modello finale...")
model.load_state_dict(torch.load('/home/jupyter-leosky01/ATP_pred/best_model_invariant.pt'))
test_loss, test_auc, test_acc, test_preds, test_labels = evaluate(
    model, test_loader, criterion, device
)

print(f"\nTest Results:")
print(f"AUC: {test_auc:.4f}")
print(f"Accuracy: {test_acc:.4f}")

# 13) Plot delle metriche
print("Creazione grafici...")
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.title('Loss over epochs')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(train_aucs, label='Train AUC')
plt.plot(val_aucs, label='Val AUC')
plt.title('AUC over epochs')
plt.legend()

plt.tight_layout()
plt.savefig('/home/jupyter-leosky01/ATP_pred/training_curves_invariant.png')
plt.show()

# 14) Analisi delle feature importance
print("Analisi feature importance...")
def get_feature_importance(model, X_val, feature_names, n_samples=1000):
    model.eval()
    X_tensor = torch.tensor(X_val, dtype=torch.float32).to(device)
    baseline_preds = torch.sigmoid(model(X_tensor)).mean().item()
    
    importances = []
    for i, name in enumerate(feature_names):
        X_perturbed = X_val.copy()
        X_perturbed[:, i] = np.random.permutation(X_perturbed[:, i])
        perturbed_preds = torch.sigmoid(model(torch.tensor(X_perturbed, dtype=torch.float32).to(device))).mean().item()
        importance = abs(baseline_preds - perturbed_preds)
        importances.append((name, importance))
    
    return sorted(importances, key=lambda x: x[1], reverse=True)

# Prendi un subset per l'analisi
feature_names = num_cols + ['Surface_' + str(i) for i in range(X_cat.shape[1] - len(num_cols))]
X_val = X[val_i]
feature_importances = get_feature_importance(model, X_val[:1000], feature_names)

print("\nTop 10 Feature Importances:")
for name, importance in feature_importances[:10]:
    print(f"{name}: {importance:.4f}")

# 15) Salva i modelli e le trasformazioni per la predizione
print("Salvataggio modelli e trasformazioni...")
joblib.dump(scaler, '/home/jupyter-leosky01/ATP_pred/scaler_invariant.joblib')
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
encoder.fit(df[['Surface','Tournament']].fillna('NA'))
joblib.dump(encoder, '/home/jupyter-leosky01/ATP_pred/cat_encoder_invariant.joblib')

# Salva anche le colonne per il preprocessing
with open('/home/jupyter-leosky01/ATP_pred/feature_cols_invariant.txt', 'w') as f:
    f.write(','.join(num_cols))

print("Addestramento completato con successo!")
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import roc_auc_score, accuracy_score, precision_recall_curve, classification_report
from sklearn.isotonic import IsotonicRegression
import matplotlib.pyplot as plt
import joblib
import warnings
from tqdm import tqdm
from collections import defaultdict
import seaborn as sns
from sklearn.metrics import roc_curve, auc

warnings.filterwarnings("ignore", category=UserWarning)

class EnhancedMLP(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.bn1 = nn.BatchNorm1d(in_dim)
        self.net = nn.Sequential(
            nn.Linear(in_dim, 512), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4929775),
            nn.BatchNorm1d(512),
            
            nn.Linear(512, 128), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4220651),
            nn.BatchNorm1d(128),
            
            nn.Linear(128, 64),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.1701431),
            
            nn.Linear(64, 1)
        )
        
    def forward(self, x):
        return self.net(self.bn1(x))

def calcola_segnali_upset(df_row):
    """
    Calcola segnali che indicano un upset potenziale legittimo
    Ritorna un punteggio 0-1 dove 1 = segnali fortissimi di upset
    """
    segnali = 0
    
    # 1. Form recente molto diversa (più peso)
    form_diff = df_row['p1_win_rate'] - df_row['p2_win_rate']
    if form_diff > 0.2:  # P1 in forma molto migliore
        segnali += 0.3
    elif form_diff > 0.1:
        segnali += 0.15
    
    # 2. Streak positivo per underdog
    if df_row['p1_streak'] > 3:  # P1 su buona striscia
        segnali += 0.2
    
    # 3. H2H favorevole
    if df_row['h2h_win_rate'] > 0.6 and df_row['h2h_matches'] >= 3:
        segnali += 0.25
    
    # 4. Odds vs ranking inconsistency (market segnala qualcosa)
    rank_ratio = df_row['Rank_2'] / max(df_row['Rank_1'], 1)  # Quanto P2 è meglio ranked
    odds_ratio = df_row['Odd_2'] / df_row['Odd_1']  # Quanto market favorisce P1
    
    # Se market è più ottimista su P1 rispetto ai ranking
    if odds_ratio < rank_ratio * 0.8:
        segnali += 0.15
    
    # 5. Surface specialty (da implementare se hai dati)
    # Se P1 è specialista della superficie...
    
    return min(segnali, 1.0)

def threshold_bilanciato(rank_diff, surface, segnali_upset, upset_bias_factor=0.85):
    """
    Threshold più intelligente che considera segnali di upset
    """
    # Base threshold conservativo
    if rank_diff <= -50:
        base = 0.22
    elif rank_diff <= -20:
        base = 0.32
    elif rank_diff <= 0:
        base = 0.47
    elif rank_diff <= 20:
        base = 0.45  # Meno conservativo di prima (era 0.65)
    elif rank_diff <= 50:
        base = 0.52  # Meno conservativo di prima (era 0.75)
    else:
        base = 0.55  # Molto meno conservativo di prima (era 0.85)
    
    # Adjustment per superficie
    surface_adj = {'Hard': 0.002, 'Clay': 0.001, 'Grass': 0.002}.get(surface, 0)
    base += surface_adj
    
    # CRUCIALE: Riduci threshold se ci sono segnali forti di upset
    if segnali_upset > 0.45:  # Segnali molto forti
        base -= 0.22
    elif segnali_upset > 0.25:  # Segnali moderati
        base -= 0.19
    elif segnali_upset > 0.1:  # Segnali leggeri
        base -= 0.12
    
    return max(base, 0.22)  # Threshold minimo per evitare troppe predizioni casuali

def ottimizza_modello_bilanciato():
    """
    Versione bilanciata: riduce falsi positivi MA cattura upset con segnali forti
    """
    
    print("=== OTTIMIZZAZIONE BILANCIATA MODELLO ATP ===\n")
    print("🎯 Obiettivo: Ridurre falsi positivi mantenendo capacità di catturare upset")
    
    # 1) Preprocessing (stesso codice di prima)
    print("Caricamento e preprocessing dati...")
    df = pd.read_csv('/home/jupyter-leosky01/ATP_pred/data/atp_tennis_cleaned.csv', parse_dates=['Date'])
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    start, end = pd.to_datetime('2010-01-01'), pd.to_datetime('2025-11-23')
    df = df[(df['Date'] >= start) & (df['Date'] <= end)].copy().sort_values('Date')
    
    # Feature engineering (stesso di prima - ommetto per brevità ma deve essere identico)
    df['p1_win'] = (df['Winner'] == df['Player_1']).astype(int)
    df['p2_win'] = (df['Winner'] == df['Player_2']).astype(int)
    
    # Statistiche mobili
    player_stats = []
    for player in tqdm(set(df['Player_1']).union(set(df['Player_2'])), desc="Elaborazione giocatori"):
        player_matches = pd.concat([
            df[df['Player_1'] == player][['Date', 'p1_win']].rename(columns={'p1_win': 'win'}),
            df[df['Player_2'] == player][['Date', 'p2_win']].rename(columns={'p2_win': 'win'})
        ]).sort_values('Date')
        
        player_matches['win_rate'] = player_matches['win'].transform(
            lambda x: x.rolling(window=10, min_periods=1).mean().shift(1)
        )
        player_matches['streak'] = player_matches['win'].groupby(
            (player_matches['win'] != player_matches['win'].shift()).cumsum()
        ).cumcount() + 1
        player_matches['streak'] = player_matches['streak'] * player_matches['win']
        
        player_stats.append(player_matches[['Date', 'win_rate', 'streak']].assign(player=player))
    
    player_stats = pd.concat(player_stats)
    
    def merge_player_stats(df_row, stats_df, player_col, date_col):
        stats = stats_df[
            (stats_df['player'] == df_row[player_col]) & 
            (stats_df['Date'] < df_row['Date'])
        ].sort_values('Date', ascending=False)
        
        if stats.empty:
            return 0.5, 0
        else:
            return stats.iloc[0]['win_rate'], stats.iloc[0]['streak']
    
    df[['p1_win_rate', 'p1_streak']] = df.apply(
        lambda x: merge_player_stats(x, player_stats, 'Player_1', 'Date'), 
        axis=1, result_type='expand'
    )
    df[['p2_win_rate', 'p2_streak']] = df.apply(
        lambda x: merge_player_stats(x, player_stats, 'Player_2', 'Date'), 
        axis=1, result_type='expand'
    )
    
    # H2H
    h2h = defaultdict(lambda: {'p1_wins': 0, 'matches': 0})
    h2h_features = []
    
    for _, row in tqdm(df.iterrows(), total=len(df), desc="Calcolo H2H"):
        p1, p2 = row['Player_1'], row['Player_2']
        key = (p1, p2)
        
        if key not in h2h:
            h2h[key] = {'p1_wins': 0, 'matches': 0}
        
        h2h_win_rate = h2h[key]['p1_wins'] / h2h[key]['matches'] if h2h[key]['matches'] > 0 else 0.5
        
        h2h_features.append({
            'h2h_win_rate': h2h_win_rate,
            'h2h_matches': h2h[key]['matches']
        })
        
        if row['Winner'] == p1:
            h2h[key]['p1_wins'] += 1
        h2h[key]['matches'] += 1
    
    df = pd.concat([df, pd.DataFrame(h2h_features)], axis=1)
    
    # Feature invarianti
    df['rank_diff'] = df['Rank_1'] - df['Rank_2']
    df['odds_ratio'] = df['Odd_1'] / df['Odd_2']
    df['win_rate_diff'] = df['p1_win_rate'] - df['p2_win_rate']
    df['streak_diff'] = df['p1_streak'] - df['p2_streak']
    df['min_rank'] = np.minimum(df['Rank_1'], df['Rank_2'])
    df['max_rank'] = np.maximum(df['Rank_1'], df['Rank_2'])
    df['min_odds'] = np.minimum(df['Odd_1'], df['Odd_2'])
    df['max_odds'] = np.maximum(df['Odd_1'], df['Odd_2'])
    df['max_win_rate'] = np.maximum(df['p1_win_rate'], df['p2_win_rate'])
    df['min_win_rate'] = np.minimum(df['p1_win_rate'], df['p2_win_rate'])
    df['max_streak'] = np.maximum(df['p1_streak'], df['p2_streak'])
    df['min_streak'] = np.minimum(df['p1_streak'], df['p2_streak'])
    
    # NUOVA FEATURE: Segnali di upset
    print("Calcolo segnali di upset...")
    df['segnali_upset'] = df.apply(calcola_segnali_upset, axis=1)
    
    # Split dati
    n = len(df)
    idx = np.arange(n)
    np.random.seed(42)
    np.random.shuffle(idx)
    t0, t1 = int(0.7*n), int(0.85*n)
    train_i, val_i, test_i = idx[:t0], idx[t0:t1], idx[t1:]
    
    num_cols = ['rank_diff', 'odds_ratio', 'win_rate_diff', 'streak_diff',
                'min_rank', 'max_rank', 'min_odds', 'max_odds',
                'max_win_rate', 'min_win_rate', 'max_streak', 'min_streak',
                'h2h_win_rate', 'h2h_matches', 'segnali_upset']  # Aggiunta nuova feature
    
    # 2) Carica modello e genera predizioni
    print("Caricamento modello...")
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    
    X_num = df[num_cols].fillna(0).values
    encoder = joblib.load('/home/jupyter-leosky01/ATP_pred/cat_encoder_invariant.joblib')
    X_cat = encoder.transform(df[['Surface','Tournament']].fillna('NA'))
    scaler = joblib.load('/home/jupyter-leosky01/ATP_pred/scaler_invariant.joblib')
    
    # NOTA: Adatta lo scaler per la nuova feature
    print("⚠️ Attenzione: Nuova feature 'segnali_upset' - potrebbe servire re-training per risultati ottimali")
    
    # Per ora usiamo il modello esistente ignorando l'ultima feature
    X_num_old = X_num[:, :-1]  # Escludi l'ultima colonna (segnali_upset)
    X_num_scaled = scaler.transform(X_num_old)
    X = np.hstack([X_num_scaled, X_cat])
    y = df['p1_win'].fillna(0).astype(int).values
    
    model = EnhancedMLP(X.shape[1]).to(device)
    model.load_state_dict(torch.load('/home/jupyter-leosky01/ATP_pred/best_model_invariant.pt'))
    model.eval()
    
    # Predizioni
    with torch.no_grad():
        X_val_tensor = torch.tensor(X[val_i], dtype=torch.float32).to(device)
        X_test_tensor = torch.tensor(X[test_i], dtype=torch.float32).to(device)
        
        val_probs = torch.sigmoid(model(X_val_tensor)).cpu().numpy().flatten()
        test_probs = torch.sigmoid(model(X_test_tensor)).cpu().numpy().flatten()
    
    y_val, y_test = y[val_i], y[test_i]
    df_val, df_test = df.iloc[val_i], df.iloc[test_i]
    
    print(f"Predizioni baseline - Val AUC: {roc_auc_score(y_val, val_probs):.4f}")
    
    # 3) CALIBRAZIONE MENO AGGRESSIVA
    print("\n=== CALIBRAZIONE BILANCIATA ===")
    
    # Bias correction meno aggressivo
    upset_bias_factor = 0.7  # Era 0.7, ora 0.85 (meno penalizzazione)
    
    # Calibrazione separata - CORRETTO: basato su QUOTE
    df_val_copy = df_val.copy()
    df_val_copy['is_upset_potential'] = df_val_copy['Odd_1'] > df_val_copy['Odd_2']  # ✅ CORRETTO
    
    isotonic_favoriti = IsotonicRegression(out_of_bounds='clip')
    isotonic_upset = IsotonicRegression(out_of_bounds='clip')
    
    mask_favoriti = ~df_val_copy['is_upset_potential']
    mask_upset = df_val_copy['is_upset_potential']
    
    if mask_favoriti.sum() > 0:
        isotonic_favoriti.fit(val_probs[mask_favoriti], y_val[mask_favoriti])
    if mask_upset.sum() > 0:
        upset_probs_corrected = val_probs[mask_upset] * upset_bias_factor
        isotonic_upset.fit(upset_probs_corrected, y_val[mask_upset])
    
    # 4) THRESHOLD BILANCIATO SU TEST
    print("\n=== APPLICAZIONE THRESHOLD BILANCIATO ===")
    
    # Applica calibrazione su test
    df_test_copy = df_test.copy()
    df_test_copy['is_upset_potential'] = df_test_copy['Odd_1'] > df_test_copy['Odd_2']
    
    test_probs_calibrated = test_probs.copy()
    mask_test_favoriti = ~df_test_copy['is_upset_potential']
    mask_test_upset = df_test_copy['is_upset_potential']
    
    if mask_test_favoriti.sum() > 0:
        test_probs_calibrated[mask_test_favoriti] = isotonic_favoriti.predict(test_probs[mask_test_favoriti])
    if mask_test_upset.sum() > 0:
        upset_probs_corrected = test_probs[mask_test_upset] * upset_bias_factor
        test_probs_calibrated[mask_test_upset] = isotonic_upset.predict(upset_probs_corrected)
    
    # Applica threshold bilanciato
    test_predictions_bilanciato = []
    for idx, (i, row) in enumerate(df_test.iterrows()):
        prob_cal = test_probs_calibrated[idx]  # ✅ USA idx invece di i
        rank_diff = row['rank_diff']
        surface = row['Surface']
        segnali = row['segnali_upset']
        
        threshold = threshold_bilanciato(rank_diff, surface, segnali, upset_bias_factor)
        prediction = 1 if prob_cal > threshold else 0
        test_predictions_bilanciato.append(prediction)
    
    # 5) CONFRONTO RISULTATI
    print("\n=== ANALISI RISULTATI BILANCIATI ===")
    
    # Baseline
    test_pred_baseline = (test_probs > 0.5).astype(int)
    test_acc_baseline = accuracy_score(y_test, test_pred_baseline)
    
    # Modello conservativo originale (ricreiamo rapidamente)
    test_pred_conservativo = []
    for i, row in df_test.iterrows():
        prob = test_probs[idx]
        rank_diff = row['rank_diff']
        surface = row['Surface']
        
        # Threshold molto conservativo originale
        if rank_diff <= -50: thresh = 0.3
        elif rank_diff <= -20: thresh = 0.4
        elif rank_diff <= 0: thresh = 0.5
        elif rank_diff <= 20: thresh = 0.75  # Era molto conservativo
        elif rank_diff <= 50: thresh = 0.85
        else: thresh = 0.9
        
        test_pred_conservativo.append(1 if prob > thresh else 0)
    
    # Bilanciato
    test_acc_bilanciato = accuracy_score(y_test, test_predictions_bilanciato)
    test_acc_conservativo = accuracy_score(y_test, test_pred_conservativo)
    
    # Analisi upset
    df_test_results = df_test.copy()
    # CORRETTO: Upset basato sulle QUOTE
    df_test_results['is_upset'] = (df_test_results['Odd_1'] > df_test_results['Odd_2']) & (y_test == 1)
    df_test_results['pred_baseline'] = test_pred_baseline
    df_test_results['pred_conservativo'] = test_pred_conservativo
    df_test_results['pred_bilanciato'] = test_predictions_bilanciato
    
    # Falsi positivi
    fp_baseline = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                   (df_test_results['pred_baseline'] == 1) & (y_test == 0)).sum()
    fp_conservativo = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                       (df_test_results['pred_conservativo'] == 1) & (y_test == 0)).sum()
    fp_bilanciato = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                     (df_test_results['pred_bilanciato'] == 1) & (y_test == 0)).sum()
    
    # Recall upset
    upset_reali = df_test_results['is_upset'].sum()
    upset_pred_baseline = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                          (df_test_results['pred_baseline'] == 1) & (y_test == 1)).sum()
    upset_pred_conservativo = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                              (df_test_results['pred_conservativo'] == 1) & (y_test == 1)).sum()
    upset_pred_bilanciato = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & 
                            (df_test_results['pred_bilanciato'] == 1) & (y_test == 1)).sum()
    
    print(f"\n📊 CONFRONTO COMPLETO:")
    print(f"{'Metrica':<20} {'Baseline':<12} {'Conservativo':<12} {'Bilanciato':<12}")
    print("-" * 60)
    print(f"{'Accuracy':<20} {test_acc_baseline:<12.4f} {test_acc_conservativo:<12.4f} {test_acc_bilanciato:<12.4f}")
    print(f"{'Falsi Positivi':<20} {fp_baseline:<12} {fp_conservativo:<12} {fp_bilanciato:<12}")
    print(f"{'Recall Upset':<20} {upset_pred_baseline/upset_reali:<12.3f} {upset_pred_conservativo/upset_reali:<12.3f} {upset_pred_bilanciato/upset_reali:<12.3f}")
    print(f"{'Upset Catturati':<20} {upset_pred_baseline:<12} {upset_pred_conservativo:<12} {upset_pred_bilanciato:<12}")
    print(f"{'Upset Totali':<20} {upset_reali:<12} {upset_reali:<12} {upset_reali:<12}")
    
    # Focus sui casi con segnali forti
    strong_signals = df_test['segnali_upset'] > 0.4
    if strong_signals.sum() > 0:
        print(f"\n🔍 ANALISI SEGNALI FORTI (>{0.4:.1f}):")
        upset_strong = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & strong_signals & (y_test == 1)).sum()
        pred_strong_bilanciato = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & strong_signals & 
                                 (df_test_results['pred_bilanciato'] == 1) & (y_test == 1)).sum()
        pred_strong_conservativo = ((df_test_results['Odd_1'] > df_test_results['Odd_2']) & strong_signals & 
                                   (df_test_results['pred_conservativo'] == 1) & (y_test == 1)).sum()
        
        print(f"Upset con segnali forti: {upset_strong}")
        print(f"Catturati da Conservativo: {pred_strong_conservativo} ({pred_strong_conservativo/max(upset_strong,1):.1%})")
        print(f"Catturati da Bilanciato: {pred_strong_bilanciato} ({pred_strong_bilanciato/max(upset_strong,1):.1%})")

    # 6) GENERA GRAFICI COMPARATIVI
    print(f"\n📊 Generazione grafici comparativi...")
    
    def genera_grafici_comparazione(test_probs, test_probs_calibrated, y_test, df_test, 
                                    test_pred_baseline, test_pred_conservativo, test_predictions_bilanciato,
                                    test_acc_baseline, test_acc_conservativo, test_acc_bilanciato,
                                    fp_baseline, fp_conservativo, fp_bilanciato,
                                    upset_pred_baseline, upset_pred_conservativo, upset_pred_bilanciato,
                                    upset_reali):
        """
        Genera grafici comparativi per l'ottimizzazione del modello ATP
        """
        
        # Setup style
        plt.style.use('default')
        sns.set_palette("husl")
        
        # Crea figura con subplots
        fig = plt.figure(figsize=(20, 15))
        
        # a. CONFRONTO ACCURACIES
        ax1 = plt.subplot(2, 3, 1)
        models = ['Baseline', 'Conservativo', 'Bilanciato']
        accuracies = [test_acc_baseline, test_acc_conservativo, test_acc_bilanciato]
        colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
        
        bars1 = ax1.bar(models, accuracies, color=colors, alpha=0.7, edgecolor='black')
        ax1.set_ylabel('Accuracy', fontsize=12)
        ax1.set_title('📊 Confronto Accuracy Modelli', fontsize=14, fontweight='bold')
        ax1.set_ylim(0.7, 0.9)
        
        # Aggiungi valori sulle barre
        for bar, acc in zip(bars1, accuracies):
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height + 0.005,
                    f'{acc:.3f}', ha='center', va='bottom', fontweight='bold')
        
        ax1.grid(True, alpha=0.3)
        
        # b. CONFRONTO FALSI POSITIVI
        ax2 = plt.subplot(2, 3, 2)
        fp_values = [fp_baseline, fp_conservativo, fp_bilanciato]
        
        bars2 = ax2.bar(models, fp_values, color=colors, alpha=0.7, edgecolor='black')
        ax2.set_ylabel('Falsi Positivi', fontsize=12)
        ax2.set_title('❌ Confronto Falsi Positivi', fontsize=14, fontweight='bold')
        
        for bar, fp in zip(bars2, fp_values):
            height = bar.get_height()
            ax2.text(bar.get_x() + bar.get_width()/2., height + 1,
                    f'{fp}', ha='center', va='bottom', fontweight='bold')
        
        ax2.grid(True, alpha=0.3)
        
        # c. CONFRONTO RECALL UPSET
        ax3 = plt.subplot(2, 3, 3)
        recall_values = [upset_pred_baseline/upset_reali, upset_pred_conservativo/upset_reali, 
                         upset_pred_bilanciato/upset_reali]
        
        bars3 = ax3.bar(models, recall_values, color=colors, alpha=0.7, edgecolor='black')
        ax3.set_ylabel('Recall Upset', fontsize=12)
        ax3.set_title('🎯 Confronto Recall Upset', fontsize=14, fontweight='bold')
        ax3.set_ylim(0, 1)
        
        for bar, recall in zip(bars3, recall_values):
            height = bar.get_height()
            ax3.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                    f'{recall:.2f}', ha='center', va='bottom', fontweight='bold')
        
        ax3.grid(True, alpha=0.3)
        
        # d. DISTRIBUZIONE PROBABILITÀ PRE/POST CALIBRAZIONE
        ax4 = plt.subplot(2, 3, 4)
        ax4.hist(test_probs, bins=30, alpha=0.6, label='Pre-calibrazione', color='red', density=True)
        ax4.hist(test_probs_calibrated, bins=30, alpha=0.6, label='Post-calibrazione', color='blue', density=True)
        ax4.set_xlabel('Probabilità', fontsize=12)
        ax4.set_ylabel('Densità', fontsize=12)
        ax4.set_title('📈 Distribuzione Probabilità', fontsize=14, fontweight='bold')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        # e. CURVE ROC
        ax5 = plt.subplot(2, 3, 5)
        
        # ROC baseline
        fpr_base, tpr_base, _ = roc_curve(y_test, test_probs)
        auc_base = auc(fpr_base, tpr_base)
        
        # ROC calibrato
        fpr_cal, tpr_cal, _ = roc_curve(y_test, test_probs_calibrated)
        auc_cal = auc(fpr_cal, tpr_cal)
        
        ax5.plot(fpr_base, tpr_base, color='red', lw=2, 
                 label=f'Baseline (AUC = {auc_base:.3f})')
        ax5.plot(fpr_cal, tpr_cal, color='blue', lw=2, 
                 label=f'Calibrato (AUC = {auc_cal:.3f})')
        ax5.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', alpha=0.5)
        
        ax5.set_xlim([0.0, 1.0])
        ax5.set_ylim([0.0, 1.05])
        ax5.set_xlabel('False Positive Rate', fontsize=12)
        ax5.set_ylabel('True Positive Rate', fontsize=12)
        ax5.set_title('📊 Curve ROC', fontsize=14, fontweight='bold')
        ax5.legend(loc="lower right")
        ax5.grid(True, alpha=0.3)
        
        # f. METRICHE RIASSUNTIVE
        ax6 = plt.subplot(2, 3, 6)
        ax6.axis('off')
        
        # Tabella riassuntiva
        metrics_data = [
            ['Metrica', 'Baseline', 'Conservativo', 'Bilanciato'],
            ['Accuracy', f'{test_acc_baseline:.3f}', f'{test_acc_conservativo:.3f}', f'{test_acc_bilanciato:.3f}'],
            ['Falsi Positivi', str(fp_baseline), str(fp_conservativo), str(fp_bilanciato)],
            ['Recall Upset', f'{upset_pred_baseline/upset_reali:.3f}', 
             f'{upset_pred_conservativo/upset_reali:.3f}', f'{upset_pred_bilanciato/upset_reali:.3f}'],
            ['Upset Catturati', f'{upset_pred_baseline}/{upset_reali}', 
             f'{upset_pred_conservativo}/{upset_reali}', f'{upset_pred_bilanciato}/{upset_reali}']
        ]
        
        table = ax6.table(cellText=metrics_data[1:], colLabels=metrics_data[0],
                          cellLoc='center', loc='center', 
                          colWidths=[0.25, 0.25, 0.25, 0.25])
        table.auto_set_font_size(False)
        table.set_fontsize(10)
        table.scale(1, 2)
        
        # Colora header
        for i in range(len(metrics_data[0])):
            table[(0, i)].set_facecolor('#4ECDC4')
            table[(0, i)].set_text_props(weight='bold')
        
        ax6.set_title('📋 Riassunto Metriche', fontsize=14, fontweight='bold', pad=20)
        
        plt.tight_layout()
        plt.savefig('/home/jupyter-leosky01/ATP_pred/confronto_modelli_atp.png', 
                    dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
        
        # GRAFICO AGGIUNTIVO: Focus sui segnali upset
        fig2, (ax7, ax8) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Distribuzione segnali upset
        ax7.hist(df_test['segnali_upset'], bins=20, alpha=0.7, color='orange', edgecolor='black')
        ax7.axvline(0.4, color='red', linestyle='--', linewidth=2, label='Soglia segnali forti (0.4)')
        ax7.set_xlabel('Segnali Upset', fontsize=12)
        ax7.set_ylabel('Frequenza', fontsize=12)
        ax7.set_title('🚨 Distribuzione Segnali Upset', fontsize=14, fontweight='bold')
        ax7.legend()
        ax7.grid(True, alpha=0.3)
        
        # Performance per livello di segnali
        signal_ranges = [(0, 0.1), (0.1, 0.25), (0.25, 0.4), (0.4, 1.0)]
        signal_labels = ['Deboli\n(0-0.1)', 'Moderati\n(0.1-0.25)', 'Forti\n(0.25-0.4)', 'Molto Forti\n(0.4+)']
        
        bilanciato_performance = []
        for min_sig, max_sig in signal_ranges:
            mask = (df_test['segnali_upset'] >= min_sig) & (df_test['segnali_upset'] < max_sig)
            if mask.sum() > 0:
                # Upset reali in questa fascia
                upset_mask = (df_test['Odd_1'] > df_test['Odd_2']) & mask
                upset_reali_fascia = (upset_mask & (y_test == 1)).sum()
                # Upset predetti correttamente dal modello bilanciato
                upset_pred_fascia = (upset_mask & 
                                   (np.array(test_predictions_bilanciato) == 1) & 
                                   (y_test == 1)).sum()
                
                performance = upset_pred_fascia / max(upset_reali_fascia, 1)
                bilanciato_performance.append(performance)
            else:
                bilanciato_performance.append(0)
        
        bars = ax8.bar(signal_labels, bilanciato_performance, color='#45B7D1', alpha=0.7, edgecolor='black')
        ax8.set_ylabel('Recall Upset', fontsize=12)
        ax8.set_title('🎯 Performance per Livello Segnali', fontsize=14, fontweight='bold')
        ax8.set_ylim(0, 1)
        
        for bar, perf in zip(bars, bilanciato_performance):
            height = bar.get_height()
            ax8.text(bar.get_x() + bar.get_width()/2., height + 0.02,
                    f'{perf:.2f}', ha='center', va='bottom', fontweight='bold')
        
        ax8.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.savefig('/home/jupyter-leosky01/ATP_pred/analisi_segnali_upset.png', 
                    dpi=300, bbox_inches='tight', facecolor='white')
        plt.show()
        
        print("✅ Grafici salvati:")
        print("   📊 /home/jupyter-leosky01/ATP_pred/confronto_modelli_atp.png")
        print("   🚨 /home/jupyter-leosky01/ATP_pred/analisi_segnali_upset.png")

    # Chiama la funzione
    genera_grafici_comparazione(
        test_probs, test_probs_calibrated, y_test, df_test,
        test_pred_baseline, test_pred_conservativo, test_predictions_bilanciato,
        test_acc_baseline, test_acc_conservativo, test_acc_bilanciato,
        fp_baseline, fp_conservativo, fp_bilanciato,
        upset_pred_baseline, upset_pred_conservativo, upset_pred_bilanciato,
        upset_reali
    )
    
    # 7) SALVA CONFIGURAZIONE BILANCIATA
    print(f"\n💾 Salvataggio configurazione bilanciata...")
    
    config_bilanciato = {
        'isotonic_favoriti': isotonic_favoriti,
        'isotonic_upset': isotonic_upset,
        'upset_bias_factor': upset_bias_factor,
        'version': 'bilanciato',
        'metrics': {
            'baseline_accuracy': test_acc_baseline,
            'conservativo_accuracy': test_acc_conservativo,
            'bilanciato_accuracy': test_acc_bilanciato,
            'baseline_fp': fp_baseline,
            'conservativo_fp': fp_conservativo,
            'bilanciato_fp': fp_bilanciato,
            'baseline_recall': upset_pred_baseline/upset_reali,
            'conservativo_recall': upset_pred_conservativo/upset_reali,
            'bilanciato_recall': upset_pred_bilanciato/upset_reali
        }
    }
    
    joblib.dump(config_bilanciato, '/home/jupyter-leosky01/ATP_pred/optimization_config_bilanciato.joblib')
    
    print(f"✅ Configurazione salvata: /home/jupyter-leosky01/ATP_pred/optimization_config_bilanciato.joblib")
    print(f"\n🎯 RISULTATO: Modello bilanciato mantiene {test_acc_bilanciato:.1%} accuracy")
    print(f"   e cattura {upset_pred_bilanciato}/{upset_reali} upset ({upset_pred_bilanciato/upset_reali:.1%} recall)")
    print(f"   con solo {fp_bilanciato} falsi positivi (vs {fp_baseline} baseline)")

if __name__ == "__main__":
    ottimizza_modello_bilanciato()
```

```python
# =============================================================================
# 🎾 PREDICTOR ATP COMPLETO - MODELLO CALIBRATO 
# =============================================================================

import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import joblib
import warnings
from datetime import datetime, timedelta
from sklearn.isotonic import IsotonicRegression
import difflib
from pathlib import Path
import sys
warnings.filterwarnings('ignore')

if '/home/jupyter-leosky01/ATP_pred' not in sys.path:
    sys.path.insert(0, '/home/jupyter-leosky01/ATP_pred')

try:
    from src.elo_tracker import EloTracker, AdvancedEloSystem
except Exception:
    EloTracker = None
    AdvancedEloSystem = None

# Modello neurale completo
class EnhancedMLP(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.bn1 = nn.BatchNorm1d(in_dim)
        self.net = nn.Sequential(
            nn.Linear(in_dim, 512), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4929775),
            nn.BatchNorm1d(512),
            nn.Linear(512, 128), 
            nn.LeakyReLU(0.1),
            nn.Dropout(0.4220651),
            nn.BatchNorm1d(128),
            nn.Linear(128, 64),
            nn.LeakyReLU(0.1),
            nn.Dropout(0.1701431),
            nn.Linear(64, 1)
        )
        
    def forward(self, x):
        return self.net(self.bn1(x))

# Calcolatore statistiche ATP
class ATPHistoryCalculator:
    def __init__(self, dataset_path):
        self.df = pd.read_csv(dataset_path)
        self.df['Date'] = pd.to_datetime(self.df['Date'])
        self.df.sort_values('Date', inplace=True)
        self.df.reset_index(drop=True, inplace=True)
        self._players = pd.Index(pd.unique(pd.concat([self.df['Player_1'], self.df['Player_2']], ignore_index=True)))
        self._players_lower = {p.lower(): p for p in self._players}
        print(f"✅ Dataset caricato: {len(self.df)} partite storiche")

    def resolve_player_name(self, name: str) -> str:
        if not isinstance(name, str) or not name.strip():
            return name
        name = name.strip()
        if name in self._players:
            return name
        key = name.lower()
        if key in self._players_lower:
            return self._players_lower[key]
        matches = difflib.get_close_matches(name, self._players.tolist(), n=1, cutoff=0.85)
        return matches[0] if matches else name
    
    def get_player_matches_before_date(self, player_name, target_date, days_lookback=1825):
        player_name = self.resolve_player_name(player_name)
        cutoff_date = target_date - timedelta(days=days_lookback)
        player_matches = self.df[
            ((self.df['Player_1'] == player_name) | (self.df['Player_2'] == player_name)) &
            (self.df['Date'] < target_date) &
            (self.df['Date'] >= cutoff_date)
        ].copy()
        return player_matches.sort_values('Date', ascending=False)
    
    def calculate_rolling_stats(self, player_name, target_date, window=10):
        recent_matches = self.get_player_matches_before_date(player_name, target_date, days_lookback=1825)
        
        if len(recent_matches) == 0:
            return {
                'win_rate': 0.5,
                'streak': 0,
                'matches_played': 0
            }
        
        last_matches = recent_matches.head(window)
        wins = 0
        
        for _, match in last_matches.iterrows():
            won = match['Winner'] == self.resolve_player_name(player_name)
            if won:
                wins += 1
        
        win_rate = wins / len(last_matches)
        streak = self._calculate_current_streak(player_name, recent_matches)
        
        return {
            'win_rate': win_rate,
            'streak': streak,
            'matches_played': len(last_matches)
        }
    
    def calculate_surface_performance(self, player_name, target_date, surface):
        surface_matches = self.get_player_matches_before_date(player_name, target_date, days_lookback=365)
        surface_matches = surface_matches[surface_matches['Surface'] == surface]
        
        if len(surface_matches) == 0:
            return {'surface_wins': 5, 'surface_matches': 10}
        
        wins = len(surface_matches[surface_matches['Winner'] == self.resolve_player_name(player_name)])
        return {'surface_wins': wins, 'surface_matches': len(surface_matches)}
    
    def _calculate_current_streak(self, player_name, recent_matches):
        if len(recent_matches) == 0:
            return 0
            
        player_name = self.resolve_player_name(player_name)
        streak = 0
        last_result = None
        
        for _, match in recent_matches.iterrows():
            won = match['Winner'] == player_name
            
            if last_result is None:
                last_result = won
                streak = 1 if won else -1
            elif won == last_result:
                if won:
                    streak += 1
                else:
                    streak -= 1
            else:
                break
                
        return max(streak, 0)
    
    def calculate_h2h_record(self, player1, player2, target_date):
        player1 = self.resolve_player_name(player1)
        player2 = self.resolve_player_name(player2)
        h2h_matches = self.df[
            (((self.df['Player_1'] == player1) & (self.df['Player_2'] == player2)) |
             ((self.df['Player_1'] == player2) & (self.df['Player_2'] == player1))) &
            (self.df['Date'] < target_date)
        ]
        
        if len(h2h_matches) == 0:
            return 0.5, 0
        
        player1_wins = len(h2h_matches[h2h_matches['Winner'] == player1])
        return player1_wins / len(h2h_matches), len(h2h_matches)

# Sistema predizione ATP completo
class PredizioneATPCompleta:
    def __init__(self):
        MODEL_PATH = '/home/jupyter-leosky01/ATP_pred/best_model_invariant.pt'
        CONFIG_PATH = '/home/jupyter-leosky01/ATP_pred/optimization_config_bilanciato.joblib'
        SCALER_PATH = '/home/jupyter-leosky01/ATP_pred/scaler_invariant.joblib'
        ENCODER_PATH = '/home/jupyter-leosky01/ATP_pred/cat_encoder_invariant.joblib'
        DATASET_PATH = '/home/jupyter-leosky01/ATP_pred/data/atp_tennis_cleaned.csv'
        
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
        self.scaler = joblib.load(SCALER_PATH)
        self.cat_encoder = joblib.load(ENCODER_PATH)
        
        num_features = 14
        cat_features = len(self.cat_encoder.get_feature_names_out())
        input_dim = num_features + cat_features
        
        self.model = EnhancedMLP(input_dim).to(self.device)
        self.model.load_state_dict(torch.load(MODEL_PATH, map_location=self.device))
        self.model.eval()
        
        self.isotonic_favoriti = None
        self.isotonic_upset = None
        self.upset_bias_factor = 1.0
        try:
            self.config = joblib.load(CONFIG_PATH)
            if isinstance(self.config, dict):
                self.isotonic_favoriti = self.config.get('isotonic_favoriti')
                self.isotonic_upset = self.config.get('isotonic_upset')
                self.upset_bias_factor = float(self.config.get('upset_bias_factor', 1.0))
        except:
            self.config = {}
        
        self._identity_calibrator = IsotonicRegression(out_of_bounds='clip')
        self._identity_calibrator.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])

        self.history_calc = ATPHistoryCalculator(DATASET_PATH)

        self.elo_tracker = None
        self.elo_state_path = Path('/home/jupyter-leosky01/ATP_pred/models/elo_tracker/elo_state.joblib')
        self.elo_csv_path = Path(DATASET_PATH)

        if EloTracker is not None and AdvancedEloSystem is not None:
            try:
                if self.elo_state_path.exists():
                    self.elo_tracker = EloTracker.load(self.elo_state_path)
                else:
                    self.elo_tracker = EloTracker()
                    self.elo_tracker.update_from_csv(self.elo_csv_path, end_date='2025-12-31')
                    self.elo_tracker.save(self.elo_state_path)

                processed = self.elo_tracker.update_from_csv(self.elo_csv_path)
                if processed > 0:
                    self.elo_tracker.save(self.elo_state_path)
            except:
                self.elo_tracker = None

        print("✅ Sistema ATP COMPLETO caricato!")
        print(f"🎯 Modello calibrato")
        print(f"🔧 Device: {self.device}")
        if self.elo_tracker is not None:
            print(f"⚡ Elo tracker attivo: {self.elo_state_path}")
        else:
            print("⚠️ Elo tracker non disponibile")
    
    def _calcola_segnali_upset(self, features):
        segnali = 0
        
        form_diff = features['win_rate_diff']
        if form_diff > 0.2:
            segnali += 0.3
        elif form_diff > 0.1:
            segnali += 0.15
        
        if features['max_streak'] > 3:
            segnali += 0.2
        
        if features['h2h_win_rate'] > 0.6 and features['h2h_matches'] >= 3:
            segnali += 0.25
        
        rank_ratio = min(features['max_rank'], 200) / max(features['min_rank'], 1)
        odds_ratio = features['max_odds'] / features['min_odds']
        
        if odds_ratio < rank_ratio * 0.8:
            segnali += 0.15
        
        return min(segnali, 1.0)
    
    def _apply_dynamic_threshold(self, prob, rank1, rank2, odds1, odds2, surface, segnali_upset):
        rank_diff = rank1 - rank2
        
        if rank_diff <= -50:
            base = 0.22
        elif rank_diff <= -20:
            base = 0.32
        elif rank_diff <= 0:
            base = 0.47
        elif rank_diff <= 20:
            base = 0.45
        elif rank_diff <= 50:
            base = 0.52
        else:
            base = 0.55
        
        surface_adj = {'Hard': 0.002, 'Clay': 0.001, 'Grass': 0.002}.get(surface, 0)
        base += surface_adj
        
        if segnali_upset > 0.45:
            base -= 0.20
        elif segnali_upset > 0.25:
            base -= 0.19
        elif segnali_upset > 0.1:
            base -= 0.12
        
        return max(base, 0.22)

    @staticmethod
    def _normalize_surface(surface: str) -> str:
        if not isinstance(surface, str):
            return 'Hard'
        s = surface.strip().title()
        if s in {'Hard', 'Clay', 'Grass', 'Carpet'}:
            return s
        return 'Hard'

    def _mlp_prob_and_features_for_row(self, row: pd.Series):
        date = pd.Timestamp(row['Date'])
        p1 = self.history_calc.resolve_player_name(str(row['Player_1']))
        p2 = self.history_calc.resolve_player_name(str(row['Player_2']))
        surface = self._normalize_surface(str(row.get('Surface', 'Hard')))
        tournament = str(row.get('Tournament', 'NA'))
        rank1 = int(row.get('Rank_1', 9999))
        rank2 = int(row.get('Rank_2', 9999))
        odds1 = float(row.get('Odd_1', 1.9))
        odds2 = float(row.get('Odd_2', 1.9))

        stats1 = self.history_calc.calculate_rolling_stats(p1, date)
        stats2 = self.history_calc.calculate_rolling_stats(p2, date)
        h2h_winrate, h2h_matches = self.history_calc.calculate_h2h_record(p1, p2, date)

        features = {
            'rank_diff': rank1 - rank2,
            'odds_ratio': odds1 / odds2,
            'win_rate_diff': stats1['win_rate'] - stats2['win_rate'],
            'streak_diff': stats1['streak'] - stats2['streak'],
            'min_rank': min(rank1, rank2),
            'max_rank': max(rank1, rank2),
            'min_odds': min(odds1, odds2),
            'max_odds': max(odds1, odds2),
            'max_win_rate': max(stats1['win_rate'], stats2['win_rate']),
            'min_win_rate': min(stats1['win_rate'], stats2['win_rate']),
            'max_streak': max(stats1['streak'], stats2['streak']),
            'min_streak': min(stats1['streak'], stats2['streak']),
            'h2h_win_rate': h2h_winrate,
            'h2h_matches': h2h_matches
        }

        segnali_upset = self._calcola_segnali_upset(features)

        feature_names = ['rank_diff', 'odds_ratio', 'win_rate_diff', 'streak_diff',
                        'min_rank', 'max_rank', 'min_odds', 'max_odds',
                        'max_win_rate', 'min_win_rate', 'max_streak', 'min_streak',
                        'h2h_win_rate', 'h2h_matches']

        X_num = np.array([[features[f] for f in feature_names]])
        X_num_scaled = self.scaler.transform(X_num)

        cat_data = pd.DataFrame([{
            'Surface': surface,
            'Tournament': tournament
        }])
        X_cat = self.cat_encoder.transform(cat_data)

        X_final = np.concatenate([X_num_scaled, X_cat], axis=1)
        X_tensor = torch.FloatTensor(X_final).to(self.device)

        with torch.no_grad():
            raw_logit = self.model(X_tensor).item()
            raw_prob = torch.sigmoid(torch.tensor(raw_logit)).item()

        is_upset_potential = odds1 > odds2
        if is_upset_potential:
            prob_for_cal = np.clip(raw_prob * self.upset_bias_factor, 0.0, 1.0)
            cal = self.isotonic_upset or self._identity_calibrator
        else:
            prob_for_cal = raw_prob
            cal = self.isotonic_favoriti or self._identity_calibrator

        try:
            calibrated_prob = float(cal.predict([prob_for_cal])[0])
        except:
            calibrated_prob = float(raw_prob)

        calibrated_prob = float(np.clip(calibrated_prob, 0.01, 0.99))

        meta = {
            'p1': p1,
            'p2': p2,
            'surface': surface,
            'tournament': tournament,
            'rank1': rank1,
            'rank2': rank2,
            'odds1': odds1,
            'odds2': odds2,
            'segnali_upset': segnali_upset,
        }

        return calibrated_prob, meta

    def optimize_alpha_for_accuracy(self, start_date='2025-01-01', end_date='2025-12-31', alphas=None, max_matches=1500, threshold_mode='dynamic'):
        if AdvancedEloSystem is None:
            raise RuntimeError('AdvancedEloSystem non disponibile')

        sd = pd.to_datetime(start_date)
        ed = pd.to_datetime(end_date)

        df = self.history_calc.df.copy()
        df['_orig_idx'] = np.arange(len(df), dtype=int)
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df = df.dropna(subset=['Date', 'Player_1', 'Player_2', 'Winner'])
        df = df.sort_values(['Date', '_orig_idx']).reset_index(drop=True)

        eval_df = df[(df['Date'] >= sd) & (df['Date'] <= ed)].copy()
        eval_df = eval_df.dropna(subset=['Rank_1', 'Rank_2', 'Odd_1', 'Odd_2', 'Surface', 'Tournament'])
        if len(eval_df) > max_matches:
            eval_df = eval_df.tail(max_matches).copy()
        eval_ids = set(eval_df['_orig_idx'].tolist())

        y_true = (eval_df['Winner'] == eval_df['Player_1']).astype(int).values

        mlp_probs = np.zeros(len(eval_df), dtype=float)
        metas = []
        for i, (_, row) in enumerate(eval_df.iterrows()):
            p, meta = self._mlp_prob_and_features_for_row(row)
            mlp_probs[i] = p
            metas.append(meta)

        elo_probs_map = {}
        elo_sys = AdvancedEloSystem()
        df_upto = df[df['Date'] <= ed].copy()

        for _, row in df_upto.iterrows():
            orig_idx = int(row['_orig_idx'])
            p1 = str(row['Player_1']).strip()
            p2 = str(row['Player_2']).strip()
            winner = str(row['Winner']).strip()
            surface = self._normalize_surface(str(row.get('Surface', 'Hard')))

            elo_p1 = float(elo_sys.get_elo(p1))
            elo_p2 = float(elo_sys.get_elo(p2))
            elo_s1 = float(elo_sys.get_elo(p1, surface))
            elo_s2 = float(elo_sys.get_elo(p2, surface))
            elo_prob_overall = float(AdvancedEloSystem.expected_outcome(elo_p1, elo_p2))
            elo_prob_surface = float(AdvancedEloSystem.expected_outcome(elo_s1, elo_s2))
            elo_prob = float(np.clip(0.35 * elo_prob_overall + 0.65 * elo_prob_surface, 0.01, 0.99))

            if orig_idx in eval_ids:
                elo_probs_map[orig_idx] = elo_prob

            loser = p1 if winner == p2 else p2
            elo_sys.update(winner, loser, surface, pd.Timestamp(row['Date']))

        elo_probs = np.array([elo_probs_map.get(int(r['_orig_idx']), 0.5) for _, r in eval_df.iterrows()], dtype=float)

        if alphas is None:
            alphas = np.linspace(0.0, 1.0, 21)

        best_alpha = None
        best_acc = -1.0
        results = []

        for a in alphas:
            final_probs = a * mlp_probs + (1 - a) * elo_probs

            if threshold_mode == 'dynamic':
                preds = []
                for i in range(len(final_probs)):
                    meta = metas[i]
                    thr = self._apply_dynamic_threshold(
                        float(final_probs[i]),
                        meta['rank1'],
                        meta['rank2'],
                        meta['odds1'],
                        meta['odds2'],
                        meta['surface'],
                        meta['segnali_upset'],
                    )
                    preds.append(1 if float(final_probs[i]) > thr else 0)
                preds = np.array(preds, dtype=int)
            else:
                preds = (final_probs > 0.5).astype(int)

            acc = float(np.mean(preds == y_true))
            results.append((float(a), acc))
            if acc > best_acc:
                best_acc = acc
                best_alpha = float(a)

        res_df = pd.DataFrame(results, columns=['alpha', 'accuracy']).sort_values('accuracy', ascending=False)
        return {'best_alpha': best_alpha, 'best_accuracy': best_acc, 'results': res_df}

    def predici_match(self, player1, player2, rank1, rank2, odds1, odds2, surface='Hard', tournament='NA', date_str='2026-01-13', mode='mlp', alpha=0.65):
        target_date = datetime.strptime(date_str, '%Y-%m-%d')

        player1 = self.history_calc.resolve_player_name(player1)
        player2 = self.history_calc.resolve_player_name(player2)
        
        print(f"\n🎾 PREDIZIONE COMPLETA: {player1} vs {player2}")
        print(f"📅 Data: {date_str}, Superficie: {surface}")
        print(f"📊 Rankings: #{rank1} vs #{rank2}")
        print(f"💰 Quote: {odds1} vs {odds2}")
        print("-" * 60)

        elo_prob = None
        if self.elo_tracker is not None and AdvancedEloSystem is not None:
            try:
                elo_p1 = float(self.elo_tracker.get_player_elo(player1))
                elo_p2 = float(self.elo_tracker.get_player_elo(player2))
                elo_s1 = float(self.elo_tracker.get_player_elo(player1, surface=surface))
                elo_s2 = float(self.elo_tracker.get_player_elo(player2, surface=surface))
                elo_prob_overall = float(AdvancedEloSystem.expected_outcome(elo_p1, elo_p2))
                elo_prob_surface = float(AdvancedEloSystem.expected_outcome(elo_s1, elo_s2))
                elo_prob = float(np.clip(0.35 * elo_prob_overall + 0.65 * elo_prob_surface, 0.01, 0.99))
                print(f"⚡ Elo overall: {elo_p1:.0f} vs {elo_p2:.0f} | P1={elo_prob_overall:.3f}")
                print(f"⚡ Elo {surface}: {elo_s1:.0f} vs {elo_s2:.0f} | P1={elo_prob_surface:.3f}")
            except:
                elo_prob = None

        stats1 = self.history_calc.calculate_rolling_stats(player1, target_date)
        stats2 = self.history_calc.calculate_rolling_stats(player2, target_date)
        surface1 = self.history_calc.calculate_surface_performance(player1, target_date, surface)
        surface2 = self.history_calc.calculate_surface_performance(player2, target_date, surface)
        h2h_winrate, h2h_matches = self.history_calc.calculate_h2h_record(player1, player2, target_date)
        
        print(f"📈 {player1}: win_rate {stats1['win_rate']:.2f}, streak: {stats1['streak']} (matches: {stats1['matches_played']})")
        print(f"📈 {player2}: win_rate {stats2['win_rate']:.2f}, streak: {stats2['streak']} (matches: {stats2['matches_played']})")
        print(f"🏟️ {player1} su {surface}: {surface1['surface_wins']}/{surface1['surface_matches']}")
        print(f"🏟️ {player2} su {surface}: {surface2['surface_wins']}/{surface2['surface_matches']}")
        print(f"🤝 H2H storico: {player1} {h2h_winrate:.1%} ({h2h_matches} match)")
        
        features = {
            'rank_diff': rank1 - rank2,
            'odds_ratio': odds1 / odds2,
            'win_rate_diff': stats1['win_rate'] - stats2['win_rate'],
            'streak_diff': stats1['streak'] - stats2['streak'],
            'min_rank': min(rank1, rank2),
            'max_rank': max(rank1, rank2),
            'min_odds': min(odds1, odds2),
            'max_odds': max(odds1, odds2),
            'max_win_rate': max(stats1['win_rate'], stats2['win_rate']),
            'min_win_rate': min(stats1['win_rate'], stats2['win_rate']),
            'max_streak': max(stats1['streak'], stats2['streak']),
            'min_streak': min(stats1['streak'], stats2['streak']),
            'h2h_win_rate': h2h_winrate,
            'h2h_matches': h2h_matches
        }
        
        segnali_upset = self._calcola_segnali_upset(features)
        print(f"🚨 Segnali upset: {segnali_upset:.3f}")
        
        feature_names = ['rank_diff', 'odds_ratio', 'win_rate_diff', 'streak_diff',
                        'min_rank', 'max_rank', 'min_odds', 'max_odds',
                        'max_win_rate', 'min_win_rate', 'max_streak', 'min_streak',
                        'h2h_win_rate', 'h2h_matches']
        
        X_num = np.array([[features[f] for f in feature_names]])
        X_num_scaled = self.scaler.transform(X_num)
        
        cat_data = pd.DataFrame([{
            'Surface': surface,
            'Tournament': tournament
        }])
        X_cat = self.cat_encoder.transform(cat_data)
        
        X_final = np.concatenate([X_num_scaled, X_cat], axis=1)
        X_tensor = torch.FloatTensor(X_final).to(self.device)
        
        with torch.no_grad():
            raw_logit = self.model(X_tensor).item()
            raw_prob = torch.sigmoid(torch.tensor(raw_logit)).item()
        
        print(f"🔧 Raw probability: {raw_prob:.3f}")

        is_upset_potential = odds1 > odds2
        if is_upset_potential:
            prob_for_cal = np.clip(raw_prob * self.upset_bias_factor, 0.0, 1.0)
            cal = self.isotonic_upset or self._identity_calibrator
        else:
            prob_for_cal = raw_prob
            cal = self.isotonic_favoriti or self._identity_calibrator

        try:
            calibrated_prob = float(cal.predict([prob_for_cal])[0])
        except:
            calibrated_prob = raw_prob
            print("⚠️ Calibrazione fallita, usando probabilità raw")

        calibrated_prob = float(np.clip(calibrated_prob, 0.01, 0.99))
        
        if mode == 'elo' and elo_prob is not None:
            final_prob = elo_prob
        elif mode == 'blend' and elo_prob is not None:
            final_prob = float(np.clip(alpha * calibrated_prob + (1 - alpha) * elo_prob, 0.01, 0.99))
        else:
            final_prob = calibrated_prob

        print(f"🔧 Calibrated probability: {calibrated_prob:.3f}")
        if elo_prob is not None:
            print(f"⚡ Elo probability: {elo_prob:.3f}")
        print(f"✅ Final probability ({mode}): {final_prob:.3f}")
        
        threshold = self._apply_dynamic_threshold(
            final_prob, rank1, rank2, odds1, odds2, surface, segnali_upset)
        
        print(f"🔧 Dynamic threshold: {threshold:.3f}")
        
        prediction = 1 if final_prob > threshold else 0
        confidence = abs(final_prob - 0.5) * 2
        
        is_upset = (odds1 > odds2 and prediction == 1) or (odds2 > odds1 and prediction == 0)
        
        print(f"\n🏆 RISULTATO FINALE:")
        print(f"🥇 VINCITORE PREDETTO: {player1 if prediction == 1 else player2}")
        print(f"📊 Probabilità {player1}: {final_prob:.1%}")
        print(f"📊 Probabilità {player2}: {1-final_prob:.1%}")
        print(f"🎯 Confidence: {confidence:.1%}")
        print(f"🚨 Upset prediction: {'SÌ' if is_upset else 'NO'}")
        
        if segnali_upset > 0.4:
            print(f"⚠️  ATTENZIONE: Segnali upset significativi rilevati!")
        
        return {
            'winner': player1 if prediction == 1 else player2,
            'player_1_prob': final_prob,
            'player_2_prob': 1 - final_prob,
            'confidence': confidence,
            'is_upset': is_upset,
            'segnali_upset': segnali_upset,
            'mlp_prob': calibrated_prob,
            'elo_prob': elo_prob
        }

# =============================================================================
# 🚀 INIZIALIZZA IL SISTEMA COMPLETO (esegui UNA SOLA VOLTA!)
# =============================================================================
print("🔄 Inizializzazione sistema ATP COMPLETO...")
predictor = PredizioneATPCompleta()
print("🎯 Sistema pronto! Usa predictor.predici_match() per predizioni accurate!")
```

```python
from sklearn.isotonic import IsotonicRegression

# --- Legacy calibrator setup (come nella versione originale) ---
try:
    _legacy_cal = None
    if hasattr(predictor, 'config') and isinstance(predictor.config, dict):
        _legacy_cal = predictor.config.get('calibrator', None)
        _opt = predictor.config.get('optimization', None)
        if isinstance(_opt, dict):
            predictor._legacy_bias_correction_factor = float(_opt.get('bias_correction_factor', 0.92))
        else:
            predictor._legacy_bias_correction_factor = 0.92
    else:
        predictor._legacy_bias_correction_factor = 0.92

    if _legacy_cal is None:
        _legacy_cal = IsotonicRegression(out_of_bounds='clip')
        _legacy_cal.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])
    predictor._legacy_calibrator = _legacy_cal
except Exception:
    predictor._legacy_bias_correction_factor = 0.92
    _legacy_cal = IsotonicRegression(out_of_bounds='clip')
    _legacy_cal.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])
    predictor._legacy_calibrator = _legacy_cal


# --- Lookback 5 anni + Momentum (esplicito) ---
_LOOKBACK_5Y_DAYS = 365 * 5


def _rolling_stats_5y_with_momentum(self, player_name, target_date, window=10):
    recent_matches = self.get_player_matches_before_date(player_name, target_date, days_lookback=_LOOKBACK_5Y_DAYS)

    if len(recent_matches) == 0:
        return {
            'win_rate': 0.5,
            'streak': 0,
            'matches_played': 0,
            'momentum': 0.5
        }

    last_matches = recent_matches.head(window)
    wins = 0

    for _, match in last_matches.iterrows():
        won = match['Winner'] == self.resolve_player_name(player_name)
        if won:
            wins += 1

    win_rate = wins / len(last_matches)
    streak = self._calculate_current_streak(player_name, recent_matches)

    # Momentum (0..1): mix forma (win_rate) + streak normalizzata
    streak_scaled = min(max(float(streak), 0.0), 10.0) / 10.0
    momentum = float(np.clip(0.7 * float(win_rate) + 0.3 * streak_scaled, 0.0, 1.0))

    return {
        'win_rate': float(win_rate),
        'streak': int(streak),
        'matches_played': int(len(last_matches)),
        'momentum': momentum
    }


def _surface_performance_5y(self, player_name, target_date, surface):
    surface_matches = self.get_player_matches_before_date(player_name, target_date, days_lookback=_LOOKBACK_5Y_DAYS)
    surface_matches = surface_matches[surface_matches['Surface'] == surface]

    if len(surface_matches) == 0:
        return {'surface_wins': 5, 'surface_matches': 10}

    wins = len(surface_matches[surface_matches['Winner'] == self.resolve_player_name(player_name)])
    return {'surface_wins': int(wins), 'surface_matches': int(len(surface_matches))}


try:
    ATPHistoryCalculator.calculate_rolling_stats = _rolling_stats_5y_with_momentum
    ATPHistoryCalculator.calculate_surface_performance = _surface_performance_5y
except Exception:
    pass


def _legacy_calcola_segnali_upset(self, features):
    segnali = 0

    form_diff = features['win_rate_diff']
    if form_diff < -0.2:
        segnali += 0.35
    elif form_diff < -0.1:
        segnali += 0.2

    if features['streak_diff'] < -3:
        segnali += 0.25
    elif features['streak_diff'] < -1:
        segnali += 0.15

    if features['rank_diff'] > 0:
        if features['h2h_win_rate'] > 0.6 and features['h2h_matches'] >= 3:
            segnali += 0.3
        elif features['h2h_win_rate'] > 0.5 and features['h2h_matches'] >= 2:
            segnali += 0.15
    else:
        if features['h2h_win_rate'] < 0.4 and features['h2h_matches'] >= 3:
            segnali += 0.3
        elif features['h2h_win_rate'] < 0.5 and features['h2h_matches'] >= 2:
            segnali += 0.15

    rank_ratio = min(features['max_rank'], 200) / max(features['min_rank'], 1)
    odds_ratio = features['max_odds'] / max(features['min_odds'], 1.01)

    if odds_ratio < rank_ratio * 0.7:
        segnali += 0.25
    elif odds_ratio < rank_ratio * 0.85:
        segnali += 0.15

    if features['max_win_rate'] > 0.7 and features['max_rank'] > features['min_rank']:
        segnali += 0.15

    return min(segnali, 1.0)


def _legacy_calculate_confidence_improved(self, calibrated_prob, raw_prob, features, segnali_upset):
    decision_confidence = abs(calibrated_prob - 0.5) * 2
    calibration_agreement = 1 - abs(raw_prob - calibrated_prob) * 2

    odds_strength = abs(np.log(features['odds_ratio'])) / 2
    rank_strength = abs(features['rank_diff']) / 100
    signal_strength = min((odds_strength + rank_strength) / 2, 1.0)

    h2h_reliability = min(features['h2h_matches'] / 10, 1.0) if features['h2h_matches'] > 0 else 0.5

    upset_penalty = 0
    if segnali_upset > 0.5:
        if abs(calibrated_prob - 0.5) < 0.15:
            upset_penalty = 0.2

    confidence = (
        decision_confidence * 0.35 +
        calibration_agreement * 0.25 +
        signal_strength * 0.20 +
        h2h_reliability * 0.20
    ) - upset_penalty

    return max(min(confidence, 1.0), 0.0)


def _legacy_calculate_upset_probability(self, features, calibrated_prob, rank1, rank2, odds1, odds2):
    if odds1 < odds2:
        base_upset = 1 - calibrated_prob
    else:
        base_upset = calibrated_prob

    segnali = self._calcola_segnali_upset(features)

    rank_factor = abs(rank1 - rank2) / 100
    if rank_factor > 1:
        rank_factor = 1

    odds_factor = abs(np.log(odds1 / odds2)) / 2
    if odds_factor > 1:
        odds_factor = 1

    upset_prob = (
        base_upset * 0.4 +
        segnali * 0.3 +
        (1 - rank_factor) * 0.15 +
        (1 - odds_factor) * 0.15
    )

    return min(max(upset_prob, 0.0), 1.0)


def _legacy_predici_match(self, player1, player2, rank1, rank2, odds1, odds2, surface='Hard', tournament='NA', date_str='2026-01-04', mode='mlp', alpha=0.65):
    target_date = datetime.strptime(date_str, '%Y-%m-%d')

    mode_n = str(mode).strip().lower()
    if mode_n == 'blended':
        mode_n = 'blend'
    if mode_n not in {'mlp', 'elo', 'blend'}:
        mode_n = 'mlp'

    player1 = self.history_calc.resolve_player_name(player1)
    player2 = self.history_calc.resolve_player_name(player2)

    print(f"\n🎾 PREDIZIONE COMPLETA: {player1} vs {player2}")
    print(f"📅 Data: {date_str}, Superficie: {surface}")
    print(f"📊 Rankings: #{rank1} vs #{rank2}")
    print(f"💰 Quote: {odds1} vs {odds2}")
    print("-" * 60)

    elo_prob = None
    if self.elo_tracker is not None and AdvancedEloSystem is not None:
        try:
            elo_p1 = float(self.elo_tracker.get_player_elo(player1))
            elo_p2 = float(self.elo_tracker.get_player_elo(player2))
            elo_s1 = float(self.elo_tracker.get_player_elo(player1, surface=surface))
            elo_s2 = float(self.elo_tracker.get_player_elo(player2, surface=surface))
            elo_prob_overall = float(AdvancedEloSystem.expected_outcome(elo_p1, elo_p2))
            elo_prob_surface = float(AdvancedEloSystem.expected_outcome(elo_s1, elo_s2))
            elo_prob = float(np.clip(0.35 * elo_prob_overall + 0.65 * elo_prob_surface, 0.01, 0.99))
            print(f"⚡ Elo overall: {elo_p1:.0f} vs {elo_p2:.0f} | P1={elo_prob_overall:.3f}")
            print(f"⚡ Elo {surface}: {elo_s1:.0f} vs {elo_s2:.0f} | P1={elo_prob_surface:.3f}")
        except Exception:
            elo_prob = None

    stats1 = self.history_calc.calculate_rolling_stats(player1, target_date)
    stats2 = self.history_calc.calculate_rolling_stats(player2, target_date)
    surface1 = self.history_calc.calculate_surface_performance(player1, target_date, surface)
    surface2 = self.history_calc.calculate_surface_performance(player2, target_date, surface)
    h2h_winrate, h2h_matches = self.history_calc.calculate_h2h_record(player1, player2, target_date)

    m1 = float(stats1.get('momentum', stats1.get('win_rate', 0.5)))
    m2 = float(stats2.get('momentum', stats2.get('win_rate', 0.5)))

    print(f"📈 {player1}: win_rate {stats1['win_rate']:.2f}, streak: {stats1['streak']} (momentum: {m1:.2f}, matches: {stats1['matches_played']})")
    print(f"📈 {player2}: win_rate {stats2['win_rate']:.2f}, streak: {stats2['streak']} (momentum: {m2:.2f}, matches: {stats2['matches_played']})")
    print(f"🏟️ {player1} su {surface}: {surface1['surface_wins']}/{surface1['surface_matches']}")
    print(f"🏟️ {player2} su {surface}: {surface2['surface_wins']}/{surface2['surface_matches']}")
    print(f"🤝 H2H storico: {player1} {h2h_winrate:.1%} ({h2h_matches} match)")

    features = {
        'rank_diff': rank1 - rank2,
        'odds_ratio': odds1 / odds2,
        'win_rate_diff': stats1['win_rate'] - stats2['win_rate'],
        'streak_diff': stats1['streak'] - stats2['streak'],
        'min_rank': min(rank1, rank2),
        'max_rank': max(rank1, rank2),
        'min_odds': min(odds1, odds2),
        'max_odds': max(odds1, odds2),
        'max_win_rate': max(stats1['win_rate'], stats2['win_rate']),
        'min_win_rate': min(stats1['win_rate'], stats2['win_rate']),
        'max_streak': max(stats1['streak'], stats2['streak']),
        'min_streak': min(stats1['streak'], stats2['streak']),
        'h2h_win_rate': h2h_winrate,
        'h2h_matches': h2h_matches
    }

    segnali_upset = self._calcola_segnali_upset(features)
    print(f"🚨 Segnali upset: {segnali_upset:.3f}")

    feature_names = ['rank_diff', 'odds_ratio', 'win_rate_diff', 'streak_diff',
                    'min_rank', 'max_rank', 'min_odds', 'max_odds',
                    'max_win_rate', 'min_win_rate', 'max_streak', 'min_streak',
                    'h2h_win_rate', 'h2h_matches']

    X_num = np.array([[features[f] for f in feature_names]])
    X_num_scaled = self.scaler.transform(X_num)

    cat_data = pd.DataFrame([{
        'Surface': surface,
        'Tournament': tournament
    }])
    X_cat = self.cat_encoder.transform(cat_data)

    X_final = np.concatenate([X_num_scaled, X_cat], axis=1)
    X_tensor = torch.FloatTensor(X_final).to(self.device)

    with torch.no_grad():
        raw_logit = self.model(X_tensor).item()
        raw_prob = torch.sigmoid(torch.tensor(raw_logit)).item()

    print(f"🔧 Raw probability: {raw_prob:.3f}")

    # --- Legacy calibration path (come prima) ---
    try:
        cal = getattr(self, '_legacy_calibrator', None)
        if cal is None:
            cal = getattr(predictor, '_legacy_calibrator', None)
        if cal is None:
            cal = IsotonicRegression(out_of_bounds='clip')
            cal.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])
        calibrated_prob = float(cal.predict([raw_prob])[0])
    except Exception:
        calibrated_prob = float(raw_prob)

    bias = float(getattr(self, '_legacy_bias_correction_factor', getattr(predictor, '_legacy_bias_correction_factor', 0.92)))
    calibrated_prob = float(np.clip(calibrated_prob * bias, 0.05, 0.95))

    # Probabilità blended disponibile anche se mode != 'blend' (utile per analisi)
    blend_prob = None
    if elo_prob is not None:
        blend_prob = float(np.clip(alpha * calibrated_prob + (1 - alpha) * elo_prob, 0.01, 0.99))

    # Final prob usa Elo/blend, ma parte dalla calibrated_prob legacy
    if mode_n == 'elo' and elo_prob is not None:
        final_prob = elo_prob
    elif mode_n == 'blend' and elo_prob is not None:
        final_prob = blend_prob
    else:
        final_prob = calibrated_prob

    print(f"🔧 Calibrated probability: {calibrated_prob:.3f}")
    if elo_prob is not None:
        print(f"⚡ Elo probability: {elo_prob:.3f}")
    if blend_prob is not None:
        print(f"🔀 Blend probability (alpha={alpha:.2f}): {blend_prob:.3f}")
    print(f"✅ Final probability ({mode_n}): {final_prob:.3f}")

    threshold = self._apply_dynamic_threshold(final_prob, rank1, rank2, odds1, odds2, surface, segnali_upset)
    print(f"🔧 Dynamic threshold: {threshold:.3f}")

    prediction = 1 if final_prob > threshold else 0

    # --- Metriche legacy (MLP-based) ---
    confidence = self._calculate_confidence_improved(calibrated_prob, raw_prob, features, segnali_upset)
    upset_probability = self._calculate_upset_probability(features, calibrated_prob, rank1, rank2, odds1, odds2)

    # --- Metriche aggiuntive: Elo-based e Blend-based ---
    confidence_elo = None
    upset_probability_elo = None
    confidence_blend = None
    upset_probability_blend = None

    if elo_prob is not None:
        # non abbiamo un "raw_prob" per Elo: usiamo raw_prob=elo_prob per non penalizzare la calibration_agreement
        confidence_elo = self._calculate_confidence_improved(elo_prob, elo_prob, features, segnali_upset)
        upset_probability_elo = self._calculate_upset_probability(features, elo_prob, rank1, rank2, odds1, odds2)

    if blend_prob is not None:
        confidence_blend = self._calculate_confidence_improved(blend_prob, blend_prob, features, segnali_upset)
        upset_probability_blend = self._calculate_upset_probability(features, blend_prob, rank1, rank2, odds1, odds2)

    is_upset = (odds1 > odds2 and prediction == 1) or (odds2 > odds1 and prediction == 0)

    print(f"\n🏆 RISULTATO FINALE:")
    print(f"🥇 VINCITORE PREDETTO: {player1 if prediction == 1 else player2}")
    print(f"📊 Probabilità {player1}: {final_prob:.1%}")
    print(f"📊 Probabilità {player2}: {1-final_prob:.1%}")
    print(f"🎯 Confidence: {confidence:.1%}")
    print(f"🚨 Upset prediction: {'SÌ' if is_upset else 'NO'}")
    print(f"📈 Upset probability: {upset_probability:.1%}")

    if segnali_upset > 0.4:
        print("⚠️  ATTENZIONE: Segnali upset significativi rilevati!")

    return {
        'winner': player1 if prediction == 1 else player2,
        'player_1_prob': final_prob,
        'player_2_prob': 1 - final_prob,
        'confidence': confidence,
        'is_upset': is_upset,
        'upset_probability': upset_probability,
        'segnali_upset': segnali_upset,
        'mlp_prob': calibrated_prob,
        'elo_prob': elo_prob,
        'blend_prob': blend_prob,
        'confidence_elo': confidence_elo,
        'upset_probability_elo': upset_probability_elo,
        'confidence_blend': confidence_blend,
        'upset_probability_blend': upset_probability_blend,
        'mode': mode_n,
        'alpha': alpha,
        'momentum_p1': m1,
        'momentum_p2': m2,
        'momentum_diff': m1 - m2,
    }


PredizioneATPCompleta._calcola_segnali_upset = _legacy_calcola_segnali_upset
PredizioneATPCompleta._calculate_confidence_improved = _legacy_calculate_confidence_improved
PredizioneATPCompleta._calculate_upset_probability = _legacy_calculate_upset_probability
PredizioneATPCompleta.predici_match = _legacy_predici_match
```

```python
from pathlib import Path
import joblib
import torch
from sklearn.isotonic import IsotonicRegression

inv_dir = Path('/home/jupyter-leosky01/ATP_pred/models/invariant')

MODEL_PATH = inv_dir / 'best_model_invariant.pt'
SCALER_PATH = inv_dir / 'scaler_invariant.joblib'
ENCODER_PATH = inv_dir / 'cat_encoder_invariant.joblib'
CONFIG_PATH = inv_dir / 'optimization_config_bilanciato.joblib'

predictor.scaler = joblib.load(SCALER_PATH)
predictor.cat_encoder = joblib.load(ENCODER_PATH)

# Ricostruisci il modello con la dimensione corretta (dipende dall'encoder)
num_features = 14
cat_features = len(predictor.cat_encoder.get_feature_names_out())
input_dim = num_features + cat_features
predictor.model = EnhancedMLP(input_dim).to(predictor.device)
predictor.model.load_state_dict(torch.load(MODEL_PATH, map_location=predictor.device))
predictor.model.eval()

# Ricarica config storica (invariant)
try:
    predictor.config = joblib.load(CONFIG_PATH)
except Exception:
    predictor.config = {}

# Mantieni compatibilità con eventuale logica favorita/upset già presente
if isinstance(predictor.config, dict):
    predictor.isotonic_favoriti = predictor.config.get('isotonic_favoriti')
    predictor.isotonic_upset = predictor.config.get('isotonic_upset')
    predictor.upset_bias_factor = float(predictor.config.get('upset_bias_factor', 1.0))

# Setup legacy calibrator + bias come nella versione originale
try:
    _legacy_cal = None
    if isinstance(predictor.config, dict):
        _legacy_cal = predictor.config.get('calibrator', None)
        _opt = predictor.config.get('optimization', None)
        if isinstance(_opt, dict):
            predictor._legacy_bias_correction_factor = float(_opt.get('bias_correction_factor', 0.92))
        else:
            predictor._legacy_bias_correction_factor = 0.92
    else:
        predictor._legacy_bias_correction_factor = 0.92

    if _legacy_cal is None:
        _legacy_cal = IsotonicRegression(out_of_bounds='clip')
        _legacy_cal.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])
    predictor._legacy_calibrator = _legacy_cal
except Exception:
    predictor._legacy_bias_correction_factor = 0.92
    _legacy_cal = IsotonicRegression(out_of_bounds='clip')
    _legacy_cal.fit([0.1, 0.5, 0.9], [0.1, 0.5, 0.9])
    predictor._legacy_calibrator = _legacy_cal

print('✅ Predictor ricaricato da models/invariant (artifact stabili)')
```
